# Bar Diagram or Bar Graph(For Qualitative/Categorical/Nominal Data)

```python
# Creating the Categorical Nominal Data
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())
  
fig = pyplot.figure(figsize = (10, 5))
 
# Creating the bar plot
pyplot.bar(courses, values, color ='maroon',
        width = 0.4)
 
pyplot.xlabel("Courses offered")
pyplot.ylabel("No. of students enrolled Frequency")
pyplot.title("Students enrolled in different courses")
pyplot.show()
```

# Pie Diagram(For Qualitative/Categorical/Nominal Data)

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
 
data = [23, 17, 35, 29, 12, 41]
 
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
 
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*numpy.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
 
# Creating plot
fig, ax = pyplot.subplots(figsize =(12, 9))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = cars,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
 
# Adding legend
ax.legend(wedges, cars,
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
pyplot.setp(autotexts, size = 9, weight ="bold")
ax.set_title("Customizing pie chart")
 
# show plot
pyplot.show()
```

# Histogram or Column Diagram Alongwith Frequency Polygon (For Quantitative/Numerical Data)

```python
# generate a sample of random gaussians
rcParams["figure.figsize"] = (15,8)
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot
# seed the random number generator
seed(1)
# generate univariate observations
data = 20 * randn(25000) + 70
# histogram of generated data
a,bins,c=pyplot.hist(data, bins=25, density=False, color='b')
l=list(bins)
l.insert(0,0)
l.insert(len(bins)+1,bins[len(bins)-1])
mid=[]
for i in range(len(l)-1):
    ele=(l[i]+l[i+1])/2
    mid.append(ele)
x=list(a)
x.insert(0,0)
x.insert(len(a)+1,0)
pyplot.plot(mid,x,'go--')
pyplot.xlabel("X axis is for Age Group Interval", fontweight ='bold')
pyplot.ylabel("Y axis is for Frequency of Each Age Group", fontweight ='bold')
pyplot.title("Histogram Frequency Distribution of Age", fontweight ='bold')
pyplot.show()
```
